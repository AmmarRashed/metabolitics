```python
from cameo.core.solver_based_model import to_solver_based_model
import pandas as pd
import cobra
import cobra.test
from cameo import flux_variability_analysis
from IPython.display import display, HTML
import time
```

```python
running_time = time.time()
cobra_model = cobra.test.create_test_model("ecoli")
cobra_fva_result = cobra.flux_analysis.flux_variability_analysis(cobra_model)
df_cobra = pd.DataFrame.from_dict(cobra_fva_result).T
print(time.time() - running_time)
```

```python
cobra_model.objective
```

```python
running_time = time.time()
cameo_model = to_solver_based_model(cobra_model)
cameo_fva_result = flux_variability_analysis(cameo_model, fraction_of_optimum=1)
df_cameo = cameo_fva_result.data_frame
print(time.time() - running_time)
```

```python
cameo_model.objective.expression
```

```python
cameo_model.solver
```

```python
df = pd.concat([df_cobra, df_cameo], axis=1, join='inner')
```

```python
df = df[(df.T != 0).any()].round(5)
```
