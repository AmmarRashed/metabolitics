```python
keys = list(set([i for i, _ in diseases_sims] + [j for _, j in diseases_sims]))
```

```python
from collections import OrderedDict
df_dict = OrderedDict()

for i in keys:
    df_dict[i] = list()
    for j in keys:
        if i == j:
            df_dict[i].append(0)
        elif (i, j) in diseases_sims:
            df_dict[i].append(diseases_sims[i, j])
        else:
            df_dict[i].append(diseases_sims[j, i])
```

```python
df_sim =  (1 / DataFrame(df_dict, index=keys)).replace([np.inf], 0)
```
