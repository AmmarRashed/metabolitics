```python
diseases_sims = {
 ('canavan disease', 'celiac disease'): 1.902,
 # ('canavan disease', 'heart failure'): 1.490,
 ('canavan disease', 'schizophrenia'): 2.288,
 # ('celiac disease', 'heart failure'): 2.034,
 ('celiac disease', 'schizophrenia'): 2.675,
 ('cirrhosis', 'canavan disease'): 1.469,
 ('cirrhosis', 'celiac disease'): 1.971 ,
 # ('cirrhosis', 'diabetes'): 1.132,
 ('cirrhosis', 'epilepsy'): 0.539,
 ('cirrhosis', 'schizophrenia'): 1.752,
 # ('diabetes', 'canavan disease'): 1.765,
 # ('diabetes', 'celiac disease'): 2.060,
 # ('diabetes', 'heart failure'): 1.683,
 ('epilepsy', 'canavan disease'): 1.822,
 ('epilepsy', 'celiac disease'): 1.561, 
 # ('epilepsy', 'diabetes'): 1.104,
 # ('heart failure', 'cirrhosis'): 1.971,
 # ('heart failure', 'epilepsy'): 0.958,
 # ('heart failure', 'schizophrenia'): 1.916,
 # ('schizophrenia', 'diabetes'): 1.888,
 ('schizophrenia', 'epilepsy'): 2.108,
}
```

```python
keys = list(set([i for i, _ in diseases_sims]))
```

```python
from collections import OrderedDict
df_dict = OrderedDict()

for i in keys:
    df_dict[i] = list()
    for j in keys:
        if i == j:
            df_dict[i].append(0)
        elif (i, j) in diseases_sims:
            df_dict[i].append(diseases_sims[i, j])
        else:
            df_dict[i].append(diseases_sims[j, i])
```

```python
df_sim =  (1 / DataFrame(df_dict, index=keys)).replace([np.inf], 0)
```

```python
from scipy.stats import kendalltau

ks = list()

for i in df_sim:
    # print(df_sim[i].sort_values(),'\n' , df_our[i].sort_values())
    k = kendalltau(df_sim[i].sort_values().keys()[1:], df_our[i].sort_values().keys()[1:])[0]
    print(i, k)
    ks.append(k)

    
print('\n', 'mean:', np.mean(ks))
```
